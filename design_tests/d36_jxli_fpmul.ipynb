## Notes
Set dip switches project selection region for design number 36, binary 0b100100

```python
c = Chip("/dev/tty.usbmodem1101", 115200)
c.step_clock(10)
```

```python
def send_number(x: int):
    print(f"Sending number {x:08b}")
    # Send high bits
    value = (0 << 7) | ((x & 0b11110000) // 2) | (0b100)
    print(f"Expected state HIGH, sending in value {value:08b}")
    c.set_all_inputs(value)
    c.step_clock()
    value = (0 << 7) | ((x & 0b11110000) // 2) | (0b101)
    print(f"Expected state HIGH, sending in value {value:08b}")
    c.set_all_inputs(value)
    c.step_clock()

    # Send low bits
    value = (0 << 7) | ((x & 0b00001111) << 3) | (0b100)
    print(f"Expected state LOW, sending in value {value:08b}")
    c.set_all_inputs(value)
    c.step_clock()
    value = (0 << 7) | ((x & 0b00001111) << 3) | (0b101)
    print(f"Expected state LOW, sending in value {value:08b}")
    c.set_all_inputs(value)
    c.step_clock()


def send_numbers(a: int, b: int):
    print("Sending a...")
    send_number(a)
    print("Sending b...")
    send_number(b)

def flush(n:int):
    # manually step the clock
    for _ in range(n):
        value = 0b0000_0000
        c.set_all_inputs(value)
        c.step_clock()
        value = 0b0000_0001
        c.set_all_inputs(value)
        c.step_clock()

def reset_manual():
    value = 0b0000_0010
    c.set_all_inputs(value)
    c.step_clock()
    value = 0b0000_0011
    c.set_all_inputs(value)
    c.step_clock()
```

```python
neg_twenty = 0b11010010
three = 0b01000100
reset_manual()
send_numbers(neg_twenty, three)
flush(100)
out = c.get_all_outputs()
print(f"0b{out:08b}")
```
